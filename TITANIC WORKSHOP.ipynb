```python
titanic_survivors=titanic_df.query('survived== 1')
titanic_survivors.count()
```

```python
titanic_no_survivors=titanic_df.query('survived== 0')
titanic_no_survivors.count()
```

```python
titanic_survivors.groupby('sex').count()
```

```python
titanic_no_survivors.groupby('sex').count()
```
